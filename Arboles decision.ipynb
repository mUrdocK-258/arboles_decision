# Arboles de decisión


```python
# importamos las librerias necesararias.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
A_decision = pd.read_csv("C:/Users/Pablo Álvarez A/Pablo Alvarez/Arboles de decisión/wine.data")
```

```python
predictors_col = ["14.23","1.71","2.43"]
target_col = ['1065']
```

```python
predictors = A_decision[predictors_col]
target = A_decision[target_col]
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=1)
```

```python
tree = DecisionTreeClassifier()
tree = tree.fit(X_train, y_train)
```

```python
plot_tree(tree)
```

```python
predicciones = tree.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predicciones)
accuracy
```
